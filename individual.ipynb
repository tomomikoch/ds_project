# Different Visualizations
##### Graph 1: Age vs Experience
##### Graph 2: Subscription vs Age
##### Graph 3: Experience vs Hours 

### Graph 1

```r
#data for graph 1

experience_and_age <- players |>
group_by(experience) |>
summarise(mean_age = mean(Age, na.rm = TRUE))

experience_and_age
```

```r
graph1 <- ggplot(experience_and_age, aes(x = experience, y = mean_age, fill = experience)) +
                 geom_bar(stat = "identity") +
                 labs(x = "Experience Level", y = "Mean Age", title = "Experience Level vs Mean Age")
graph1 
```

### Graph 2

```r
# data for graph 2 
subscribe_and_age <- players |>
group_by(subscribe) |>
summarise(mean_age = mean(Age, na.rm = TRUE))

subscribe_and_age
```

```r
graph2 <- ggplot(subscribe_and_age, aes(x = subscribe, y = mean_age, fill = subscribe)) +
                 geom_bar(stat = "identity") +
                 labs(x = "Whether or not subscribed", y = "Mean Age", title = "Subscription status vs Mean Age")
graph2 
```

### Graph 3

```r
#data for graph 3

experience_and_hours <- players |>
group_by(experience) |>
summarise(mean_hours = mean(played_hours, na.rm = TRUE))

experience_and_hours
```

```r
graph3 <- ggplot(experience_and_hours, aes(x = experience, y = mean_hours, fill = experience)) +
                 geom_bar(stat = "identity") +
                 labs(x = "Experience Level", y = "Average Hours", title = "Experience Level vs Average Hours")

graph3
```

## Analyzation of each graph

**Graph 1:** In graph 1, I compared experience level vs mean age. The lowest mean age was for the experience level "pro", at around 17 years, and the highest mean age was for the "beginner" experience level at around 21 years old. This graph is relevant to my research question because it shows me the avaerage age difference in each group. If assuming that the more someone plays, the higher experience level they are, we can assume that the "pro" level has been playing since they were young. 


**Graph 2:** In graph 2, I compared subscription status vs mean age. The average age for people who were not subscribed was higher at around 23 years old, compared to the subscribed average age, which was around 19 years old. This graph could be relevant to my research question because it tells me that younger audiences could be more likely to subscribe. 

**Graph 3:** In graph 3, I compared experience level vs average hours played. Experience level "regular" had the highest average playing time at about 18.2 hours. Interestingly, experience level "veteran" had the lowest average playing time at about 0.6 hours. This graph is not directly relevant to my research question but is good background information.
